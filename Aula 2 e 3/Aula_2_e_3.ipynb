### Inserção de colunas no dataframe de estudo
Aqui, irei deixar de lado o dataframe principal e criar um dataframe de estudo, mas agora sem algumas das colunas que fizemos de exercício no Google Sheet, na aula anterior.

#### Criação do df_estudo

```python
# Fazendo cópia das 4 primeiras colunas do df_principal e inserindo os dados em df_estudo
df_estudo = df_principal.iloc[:,:4].copy()
df_estudo.head(5)
```

#### Alterando colunas
O Python tem a capacidade de distinguir entre letras maiúsculas e minúsculas, acentos, e reconhecer outros caracteres especiais. Aqui, estarei renomeando as colunas existentes para tornar o processo de análise mais simples.

```python
# Renomeando as colunas
df_estudo = df_estudo.rename(columns={
'Ativo': 'ativo',
'Data': 'data',
'Último (R$)':'valor_final',
'Var. Dia (%)':'var_dia_pct'
})
# Visualizando mudanças
df_estudo
```

#### Criando colunas de variação percentual e valor inicial do ativo

```python
df_estudo['var_pct'] = df_estudo['var_dia_pct'] / 100
df_estudo['valor_inicial'] = df_estudo['valor_final'] / (df_estudo['var_pct'] + 1)
df_estudo
```

#### Mesclando dataframes para descobrir a quantidade teórica de ações
 df_estudo X df_total_de_acoes




```python
# Mesclando df_estudo com df_total_de_acoes
df_estudo = df_estudo.merge(df_total_de_acoes, left_on='ativo', right_on='Código', how='left')
df_estudo
```

```python
# Limpeza do novo dataframe gerado
df_estudo = df_estudo.drop(columns=['Código'])
df_estudo.rename(columns={'Qtde. Teórica': 'qtd_teorica'}, inplace = True)
df_estudo
```

#### Criando coluna de cálculo da variação total em reais

```python
df_estudo['variacao_rs'] = (df_estudo['valor_final'] - df_estudo['valor_inicial'])*df_estudo['qtd_teorica']
df_estudo
```

```python
# Formataremos a exibição das variáveis tipo "float" porque a exibição atual está comprometendo o entendimento do conteúdo
pd.options.display.float_format = '{:.2f}'.format
# Aproveitando o embalo, por ser uma informação que não cabe valores flutuantes, converteremos a quantidade teórica para o tipo inteiro.
df_estudo['qtd_teorica'] = df_estudo['qtd_teorica'].astype(int)
df_estudo
```

#### Criando coluna para exibir o resultado da variação

```python
df_estudo['resultado'] = df_estudo['variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_estudo
```

#### Mesclando dataframes para encontrar o nome da empresa referenciada por determinado ativo

df_estudo X df_ticker

```python
df_estudo = df_estudo.merge(df_ticker, left_on='ativo', right_on='Ticker', how='left')
```

```python
df_estudo = df_estudo.drop(columns=['Ticker'])
df_estudo.rename(columns={'Nome': 'nome'}, inplace=True)
df_estudo
```

#### Mesclando dataframes para inserção das colunas "segmento" e "idade_em_anos"
df_estudo X df_chatgpt

```python
df_estudo = df_estudo.merge(df_chatgpt, left_on='nome', right_on='Nome da Empresa', how='left')
```

```python
df_estudo = df_estudo.drop(columns=['Nome da Empresa'])
df_estudo = df_estudo.rename(columns={
'Segmento':'segmento',
'Idade (em anos)':'idade_em_anos'
})
df_estudo
```

####Criação da coluna cat_idade

```python
df_estudo['cat_idade'] = df_estudo['idade_em_anos'].apply(lambda x: 'Mais de 100' if x > 100 else ('Menos de 50' if x < 50 else 'Entre 50 e 100'))
df_estudo
```

###Análises

```python
# Calculando o maior valor
maior = df_estudo['variacao_rs'].max()

# Calculando o menor valor
menor = df_estudo['variacao_rs'].min()

# Calculando a média
media = df_estudo['variacao_rs'].mean()

# Calculando a média de quem subiu
media_subiu = df_estudo[df_estudo['resultado'] == 'Subiu']['variacao_rs'].mean()

# Calculando a média de quem desceu
media_desceu = df_estudo[df_estudo['resultado'] == 'Desceu']['variacao_rs'].mean()

# Imprimindo os resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
# Exibindo os 5 primeiros ativos que tiveram a maior variação percentual dentre os demais
df_estudo_subiu = df_estudo[df_estudo['resultado'] == 'Subiu']
df_estudo_subiu.head(5)
```

```python
df_analise_segmento = df_estudo_subiu.groupby('segmento')['variacao_rs'].sum().reset_index()
df_analise_segmento = df_analise_segmento.sort_values(by='variacao_rs', ascending=False)
df_analise_segmento = df_analise_segmento.reset_index(drop=True)
df_analise_segmento
```

```python
df_analise_saldo = df_estudo.groupby('resultado')['variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
# Você pode excluir a linha onde 'resultado' é 'Estável' usando a função drop
df_analise_saldo = df_analise_saldo.drop(df_analise_saldo[df_analise_saldo['resultado'] == 'Estável'].index)

# Agora, ordene o DataFrame pelo maior valor de 'variacao_rs'
df_analise_saldo = df_analise_saldo.sort_values(by='variacao_rs', ascending=False)

# Agora, resete o índice do DataFrame
df_analise_saldo = df_analise_saldo.reset_index(drop=True)

df_analise_saldo
```

```python
# Função para definir as cores com base em seu resultado
def cor_grafico(resultado):
    if resultado == 'Subiu':
        return 'Subiu'
    else:
        return 'Desceu'

# Inserindo coluna de cores para plotagem de gráfico
df_analise_saldo_grafico = df_analise_saldo.copy()
df_analise_saldo_grafico['cor'] = df_analise_saldo_grafico['resultado'].apply(cor_grafico)

# Criação do gráfico de barras com cores condicionais
fig = px.bar(df_analise_saldo_grafico,
             x='resultado',
             y='variacao_rs',
             text='variacao_rs',
             title='Variação do fechamento por resultado'.upper(),
             color='cor',
             color_discrete_map={'Subiu': '#A3DE89', 'Desceu': '#AF3838'})

fig.update_layout(
    width=600,
    xaxis_title='Resultado'.upper(),  # Legenda do eixo x
    yaxis_title='Variação (em reais)'.upper(),  # Legenda do eixo y
    font_size= 14,
    title_x=0.5,
    title_y=0.9,
    legend=dict(
        title='Resultado',
        bgcolor="#E7E7E7",
        bordercolor='black',
        borderwidth=1,
        x=1.02,
        y=1

    )
)
# Exiba o gráfico
fig.show()
```
